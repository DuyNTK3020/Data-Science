# Data Visualization-California Cities

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
cities = pd.read_csv('./california_cities.csv')
cities.head()

cities['area_total_km2'].sort_values(ascending = False)
```

```python
# Extract Latd = Vĩ độ and Longd = Kinh độ

lat, lon = cities['latd'], cities['longd']

population, area = cities['population_total'], cities['area_total_km2']

plt.style.use('ggplot')
plt.figure(figsize = (8,6))
# Plot using Pylot API
plt.scatter(lon, lat, c = np.log10(population), s = area, alpha = 0.5);
plt.axis('equal');

plt.title('California Cities: Popularion and Area Distribution')
plt.xlabel('longlatitude')
plt.ylabel('latitude')

plt.colorbar(label = 'log$_{10}$(population)')
plt.clim(3,7)

# Create a legend for cities'size
area_range = [50, 100 ,300, 500]

for area in area_range:
    plt.scatter([], [], s = area, alpha = 0.4,
               label = str(area) + ' km$^2$')
plt.legend(labelspacing = 1)
```
